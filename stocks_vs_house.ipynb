```python
import matplotlib.pyplot as plt

from portfolio import Portfolio

years_on_loan = 30
n_months = years_on_loan * 12
n_months = 120

capital = 50000
starting_year = 2022
monthly_investment = 5000
portfolio = Portfolio(
    capital, 
    starting_year, 
    monthly_investment=monthly_investment,
)

house_price = 150000
closing_costs = 15000
loan_apr = 7
property_tax_frac = 0.01
monthly_fees = 600
monthly_rent_collected = 1500
down_payment_fraction = 0.20
portfolio.buy_a_house(
    house_price,
    closing_costs,
    loan_apr,
    property_tax_frac,
    monthly_fees,
    monthly_rent_collected,
    down_payment_fraction=down_payment_fraction,
    years_on_loan=years_on_loan,
)
net_worth_history1 = [capital]
for month in range(n_months):
    portfolio.increment_one_month()
    net_worth_history1.append(portfolio.get_net_worth())
    if portfolio.stocks.amount_invested > capital:
        portfolio.buy_a_house(
            house_price,
            closing_costs,
            loan_apr,
            property_tax_frac,
            monthly_fees,
            monthly_rent_collected,
            down_payment_fraction=down_payment_fraction,
            years_on_loan=years_on_loan,
        )
        print('number of houses:', len(portfolio.houses))


portfolio = Portfolio(
    capital, 
    starting_year, 
    monthly_investment=monthly_investment,
)
net_worth_history2 = [capital]
for month in range(n_months):
    portfolio.increment_one_month()
    net_worth_history2.append(portfolio.get_net_worth())

    
portfolio = Portfolio(
    capital, 
    starting_year, 
    monthly_investment=monthly_investment,
)
cash = capital
net_worth_history4 = [cash]
net_worth_history3 = [capital]
for month in range(n_months):
    portfolio.increment_one_month()
    net_worth_history3.append(portfolio.get_net_worth(adjust_for_inflation=False))
    cash += monthly_investment
    net_worth_history4.append(cash)
```

```python
plt.plot(net_worth_history1, label='net_worth_history_house')
plt.plot(net_worth_history2, label='net_worth_history_stocks')
plt.plot(net_worth_history3, label='net_worth_history_stocks no inflation adjustment')
plt.plot(net_worth_history4, label='cash')
plt.legend()
plt.show()
```
